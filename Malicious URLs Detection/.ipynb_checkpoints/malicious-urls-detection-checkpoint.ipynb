## Project overview 

The project consisted to evaluate different classification models to predict malicious and benign websites, based on application layer and network characteristics.  

## Problem statement 

Confidentiality and Privacy is the biggest challenge faced. If a user access the 

malicious website through a search engine with no perception, malicious scripts 

usually launch attacks to install rogue program, steal personal identities and 

credentials, so our Problem is to develop an algorithm which accurately predicts 

the malicious and the benign URLs. 

This is a binary classification supervised problem that takes the URL as an input and returns 

whether it contains malicious content or not as an output. 

We've got our dataset from Kaggle platform: https://www.kaggle.com/xwolf12/malicious-and-benign-websites 

- Data set characteristics: Multivariate, Time-series 
- Attribute characteristics: Integer, real, characters 
- Associated tasks: Classification 
- Number of instances: 1782 
- Number of attributes: 21 
- Data format: csv 

The data is imbalanced and contains 1782 labeled URL: 
- 216 malicious URLs (12 %) 

in This project I'll evaluate different supervised algorithms to classify the URLs based on this data and the features related to each URL like URL length and the number of special characters in the URL.  

## 1- Exploring the Data

### Importing the data and the neccessary libraries


```python
import numpy as np
import pandas as pd
from time import time
from matplotlib import pyplot as plt
from IPython.display import display
import seaborn as sns
%matplotlib inline
```

### Loding the data

```python
data = pd.read_csv("dataset.csv")
display(data)
```

### Statistic information about the data

```python
n_records = data['URL'].count()
n_malicious = data[data['Type']==1].shape[0]
n_benign = data[data['Type']==0].shape[0]
malicious_precentage = float(n_malicious)*100/n_records
benign_precentage = float(n_benign)*100/n_records

# Print the results
print("Total number of records: {}".format(n_records))
print("Number of malicious URLs: {}".format(n_malicious))
print("Number of benign URLs: {}".format(n_benign))
print("Percentage of malicious URLs: {} %".format(round(malicious_precentage,3)))
print("Percentage of benign URLs: {} %".format(round(benign_precentage,3)))
```

```python
data.describe(include='all')
```

From above, we see that URL is totally unique. so, we will not consider it as a feature


```python
data.drop('URL', axis=1, inplace=True) #url is unique we will not use it as a feature
data.fillna(0, inplace=True) #replacing NaN with zeros
display(data)
```

Take a look at any null values to clean up data

```python
print(data.isnull().sum())
data[pd.isnull(data).any(axis=1)]
```

The data seems to be clean so we will move to the next step

we will Convert columns into "dummy columns" so that our categorical data becomes discrete for model creation. that makes each entry have an integer unique value

```python
data = pd.get_dummies(data,prefix_sep='--')
display(data)
```

## Exploratory Visualization


## Balance the data using SMOTE algorithm

This is an important step as the ML algorithms works better with a balanced data. 

```python
X = data.drop(['Type'], axis=1)
y = data['Type']
```

```python
from imblearn.over_sampling import SMOTE

ros = SMOTE(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y)
print(X_resampled.shape)
print(y_resampled)
```

```python
import collections
collections.Counter(y_resampled)
```

Now we have equal data category 1565 for malicious URLs and 1565 for benign.

next step is to split the data into train (70%) /  test (30%) 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)
X_train_orig, X_test_orig, y_train_orig, y_test_orig = train_test_split(X, y, test_size=0.3, random_state=42)
print('X_train: ',X_train.shape,'X_test: ', X_test.shape)
print('y_train: ',y_train.shape,'y_test: ', y_test.shape)
print('X_train_orig: ',X_train_orig.shape,'X_test_orig: ', X_test_orig.shape)
print('y_train_orig: ',y_train_orig.shape,'y_test_orig: ', y_test_orig.shape)
```

### Helper functions
- print_score
    - takes the classifier and print the score.
    - if the data is imbalanced the score will contain only the recall metric. 


```python
from sklearn.model_selection import cross_val_predict, cross_val_score
from sklearn.metrics import confusion_matrix,accuracy_score, recall_score

# Visualize our results
def print_score(classifier, imbalanced_data = False):
    if imbalanced_data == True:
        train_score = recall_score(y_train_orig,classifier.predict(X_train_orig))
        test_score = recall_score(y_test_orig,classifier.predict(X_test_orig))
        print("Imbalanced data:\n")
        print("Training results:\n")
        print('Recall Score: {0:.4f}\n'.format(train_score))
        print("Testing results:\n")
        print('Recall Score: {0:.4f}\n'.format(test_score))
        print('Confusion Matrix:\n{}\n'.format(confusion_matrix(y_test,classifier.predict(X_test))))
        return {'training Recall':train_score,'testing Recall':test_score}
    train_score = accuracy_score(y_train,classifier.predict(X_train))
    train_recall = recall_score(y_train,classifier.predict(X_train))
    test_score = accuracy_score(y_test,classifier.predict(X_test))
    test_recall = recall_score(y_test,classifier.predict(X_test))
    print("Balanced data:\n")
    print("Training results:\n")
    print('Accuracy Score: {0:.4f}\n'.format(train_score))
    print("Testing results:\n")
    print('Accuracy Score: {0:.4f}\n'.format(test_score))
    print('Confusion Matrix:\n{}\n'.format(confusion_matrix(y_test,classifier.predict(X_test))))
    return {'training Recall':train_recall,'training accuracy':train_score,'testing Recall':test_recall,'testing accuracy':test_score}
    
```

## Bench mark model

```python
results = {}
results_orig = {}
```

```python
# Imbalanced data
from sklearn.dummy import DummyClassifier
dc = DummyClassifier(random_state=42)
dc.fit(X_train_orig, y_train_orig)
y_pred = dc.predict(X_test_orig)
dc_score = print_score(dc, imbalanced_data=True)
results_orig['Dummy Classifier'] = dc_score
```

```python
# Balanced data
from sklearn.dummy import DummyClassifier
dc = DummyClassifier(random_state=42)
dc.fit(X_train, y_train)
y_pred = dc.predict(X_test)
dc_score = print_score(dc)
results['Dummy Classifier'] = dc_score
```

### Training using diffrent algorithms

## Imbalanced Data

```python
from sklearn.svm import SVC
svc = SVC(gamma='auto')
svc.fit(X_train_orig, y_train_orig)
y_pred_orig = svc.predict(X_test_orig)
svc_score_orig = print_score(svc, imbalanced_data=True)
results_orig['SVC'] = svc_score_orig
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=10, max_depth=50, criterion = 'entropy')
rf.fit(X_train_orig, y_train_orig)
y_pred = rf.predict(X_test_orig)
rf_score = print_score(rf, imbalanced_data=True)
results_orig['Random Forest'] = rf_score
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train_orig, y_train_orig)
y_pred = dtc.predict(X_test_orig)
dtc_score = print_score(dtc, imbalanced_data=True)
results_orig['Decision Tree'] = dtc_score
```

## Balanced data

```python
from sklearn.svm import SVC
svc = SVC(gamma='auto')
svc.fit(X_train, y_train)
y_pred = dtc.predict(X_test)
svc_score = print_score(svc)
results['SVC'] = svc_score
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=10, max_depth=15, criterion = 'entropy')
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
rf_score = print_score(rf)
results['Random Forest'] = rf_score
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
y_pred = dtc.predict(X_test)
dtc_score = print_score(dtc)
results['Decision Tree'] = dtc_score
```

## Improvements
improve the best fit algorithm using grid search technique to find out the best parameters

```python
from sklearn.model_selection import GridSearchCV

parameters = {'criterion': ['gini', 'entropy'],'max_depth': [2,5,10,15,None]}

sg = GridSearchCV(dtc, param_grid=parameters, scoring='recall', cv=10)
grid_fit = sg.fit(X_train, y_train)
best_clf = grid_fit.best_estimator_
best_predictions = best_clf.predict(X_test)
```

```python
print(sg.best_params_)
```

```python
# default and optimized models accuracies
print("Decision Tree Classifier Results:\n")
print("Unoptimized Model Testing Recall: {:.4f}".format(results['Decision Tree']['testing Recall']))
print("Unoptimized Model Testing Accuracy: {:.4f}".format(results['Decision Tree']['testing accuracy']))

print("Optimized Model Testing Recall: {:.4f}".format(recall_score(y_test, best_predictions)))
print("Optimized Model Testing Accuracy: {:.4f}".format(accuracy_score(y_test, best_predictions)))
```
